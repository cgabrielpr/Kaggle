```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from imblearn.over_sampling._smote.base import SMOTE

%matplotlib inline
from pycaret.classification import *
from pycaret.datasets import get_data 
```

```python
# Listar todos los datasets disponibles y sus caracteristicas
all_datasets = get_data('index')
all_datasets
```

```python
cancer = get_data('cancer')
print(cancer.shape)
```

```python
columns = cancer.columns
print(columns)
```

## Modelos con Robust_Scaler

```python
datos_robust_scaler = preprocessing.RobustScaler().fit_transform(cancer)
datos_robust_scaler = pd.DataFrame(datos_robust_scaler,columns=columns)
print(datos_robust_scaler.shape)
datos_robust_scaler.head()
```

```python
sns.countplot(data = datos_robust_scaler, x='Class')
plt.show()
```

```python
ctr1 = setup(data = datos_robust_scaler,
             target='Class',
             train_size = .7,
             transformation_method = 'yeo-johnson',
             pca = False,
             remove_outliers = False,
             fold = 10,
             session_id = 2403
             )
```

```python
compare_models(sort='Accuracy')
```

```python
svm = create_model('svm')
```

```python
tuned_svm = tune_model(svm)
```

```python
plot_model(tuned_svm, plot = 'parameter')
```

```python
plot_model(tuned_svm, plot='confusion_matrix')
```

```python
evaluate_model(tuned_svm)
```

```python
model_svm = predict_model(tuned_svm)
```

## Modelos con SMOTE

```python
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(datos_robust_scaler.iloc[:,1:], datos_robust_scaler['Class'])

print('Original dataset shape', len(datos_robust_scaler))
print('Resampled dataset shape', len(y_smote))
```

```python
balance_data = pd.DataFrame(x_smote)
balance_data.head()
```

```python
balance_data['Class'] = y_smote
```

```python
sns.countplot('Class', data = balance_data)
```

```python
ctr2 = setup(data = balance_data,
             target='Class',
             train_size = .7,
             transformation_method = 'yeo-johnson',
             pca = False,
             remove_outliers = False,
             fold = 10,
             session_id = 2403
             )
```

```python
compare_models(sort = 'Accuracy')
```

```python
 et = create_model('et')
```

```python
tuned_et = tune_model(et,
                      custom_grid = {'max_features': [5,8,10,15,20,25,30],
                                     'max_depth': [1,2,4,6,8,10],
                                     'min_samples_leaf': [1,2,4,6,10,],
                                     'min_samples_split': [2,4,6,10,15]                                    
                                     },  
                      )
```

```python
plot_model(tuned_et, plot='confusion_matrix')
```

```python
plot_model(tuned_et, plot='auc')
```

```python
plot_model(tuned_et, plot = 'feature')
```

```python
Model_et = predict_model(tuned_et)
```
